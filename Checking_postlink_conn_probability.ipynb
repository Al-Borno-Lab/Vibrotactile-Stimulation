# Connection probability for each pair-wise neuron

Kromer, J. A., Khaledi-Nasab, A., & Tass, P. A. (2020). Impact of number of stimulation sites on long-lasting desynchronization effects of coordinated reset stimulation. Chaos (Woodbury, N.Y.), 30(8), 083134. https://doi.org/10.1063/5.0015196

This paper stated a connection probability based on the distance between each neuron pair (Section V subsection A).

Here I just injest the postlink hardcoded file and calcualate the mean/std/z-score of each row.

```python
import numpy as np
import pandas as pd

with open("./original_paper_fortran_code/net_reg/1_N_200_reg_postlink.net", "r") as file: 
    df = pd.read_table(file, sep="\s+", header=None)
df
```

```python
def print_mean_std(series:pd.Series):
    # return(f"Mean: {str(round(series.mean(), 5)).ljust(10)}Std: {str(round(series.std(), 5)).ljust(10)}")
    return(f"z-value: {str(round((series.mean()-series.name)/series.std(), 5))}")

result = df.apply(print_mean_std, axis=1)

with pd.option_context("display.max_rows", None):
    print(result)
```
